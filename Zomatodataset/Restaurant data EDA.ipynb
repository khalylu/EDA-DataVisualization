```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
df = pd.read_csv("zomato.csv", encoding='latin-1')
df.head()
```

```python
df.info()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.isnull(), cbar=False,cmap='Accent_r')
plt.show()
```

```python
country = pd.read_excel('Country-Code.xlsx')
country.head()
```

```python
country.info()
```

```python
usage_info = pd.merge(df, country, how='left', on='Country Code')
usage_info.head()
```

```python
cname = usage_info['Country'].value_counts().index
cname
```

```python
cval = usage_info['Country'].value_counts().values
cval
```

```python
#pie chart
plt.figure(figsize=(15,10))
plt.pie( cval[:3],labels=cname[:3], autopct='%.3f')
plt.show()
```

Overwhelming number of users are from India (94%), second is the US by 4% and UK and other countries represent less than one percent of all users.

```python
usage_info.columns
```

```python
rating = df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating counts'})
rating
```

1. 1.8 - 2.4 = poor
2. 2.5 - 3.4 = average
3. 3.5 - 3.9 = good
4. 4 - 4.4 = very good
5. 4.5 - 4.9 = excellent

```python
plt.figure(figsize=(20,15))
sns.barplot(x=rating['Aggregate rating'],y=rating['Rating counts'],hue=rating['Rating color'], 
palette=['white', 'red', 'orange', 'yellow', 'blue', 'green'])
plt.show()
```
