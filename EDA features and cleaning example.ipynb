normalization and standardization data to remove skewness

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
x = iris.data
y = iris.target
```

```python
# normalizing the features
x= MinMaxScaler().fit_transform(x)
```

```python
# same can be done for standard scaling
```

```python
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
columns= iris['feature_names'] + ['species'])
```

```python
# replace categories with species names

df['species'] = df['species'].map({0:'setosa',1:'versicolor',2:'virginica'})
```

EDA

```python
sns.countplot(x='species',data=df)
```

```python
df.hist()
```

```python
df.boxplot()
```

```python
#group categories by species names

iris = df.groupby('species').mean()
```

```python
iris.plot.bar(figsize = (12,8), fontsize = 15)
title('Average Petal and Sepal Measurements')
xlabel('Species')
ylabel('Measurements in (cm)')
grid(True)
legend(loc = 0)
xticks(rotation = 0)
```

```python
sns.heatmap(df.corr(), annot = True)
suptitle('Pair Plot', fontsize= 20)
```

```python
sns.pairplot(df, hue = 'species')
```

Final Report

There are three species of flowers Setosa, versicolor and virginica.
Average Petal width for setosa is less than one centimeter, versicolor is between 1-2 and virginica is above 2cm.
Average Petal lenght for setosa is 0-1cm, versicolor is between 2-6cm and virginica is above 4cm.

Sepal conclusion
Setosa has a sepal width between 3-4.5cm and length can reach up to 8cm, significantly longer than versicolor and virginica.

Compared to the average sepal width for versicolor and virginica is between 2.5-3.5cm and sepal lenght for versicolor and virginica can only reach 6.5cm.
